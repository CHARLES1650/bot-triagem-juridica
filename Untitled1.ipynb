```python
from openai import OpenAI

client = OpenAI()

interaction_count = 0
MAX_INTERACTIONS = 3

system_prompt = """
Você é um assistente jurídico especializado em triagem inicial de casos.

Sua função é:
- Identificar a área do Direito
- Sugerir medidas jurídicas
- Indicar documentos necessários
- Avaliar urgência

Regras:
- Não inventar informações
- Não dar valores exatos
- Máximo de 3 respostas

Na terceira resposta:
- Fazer resumo e encerrar
"""

messages = [{"role": "system", "content": system_prompt}]

def perguntar(pergunta):
    global interaction_count

    if interaction_count >= MAX_INTERACTIONS:
        print("Atendimento encerrado.")
        return

    messages.append({"role": "user", "content": pergunta})

    response = client.chat.completions.create(
        model="gpt-4.1-mini",
        messages=messages,
        temperature=0
    )

    resposta = response.choices[0].message.content
    messages.append({"role": "assistant", "content": resposta})

    interaction_count += 1

    print("\n🤖 Bot:\n")
    print(resposta)

    if interaction_count == MAX_INTERACTIONS:
        print("\n🔒 Atendimento encerrado.")
```
